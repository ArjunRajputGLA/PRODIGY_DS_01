Task-1 : Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population. 

```python
import pandas as pd
import matplotlib.pyplot as plt

file2 = "File2.csv"
df_file2 = pd.read_csv(file2) 

X = df_file2[['Country Code', 'IncomeGroup']] 
y = df_file2['IncomeGroup'].value_counts()
print(y)

plt.figure(figsize=(10, 6), dpi = 200)
y.plot(kind='bar', color='lime')
plt.title('\nDistribution of Countries by Income Group\n', fontweight='bold', fontsize = 23)
plt.xlabel('\nIncome Group\n', fontweight='bold', fontsize = 12)
plt.ylabel('Number of Countries', rotation = 40, labelpad = 55, fontweight='bold', fontsize = 10)
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```
